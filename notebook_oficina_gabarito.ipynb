##### Tarefa: Crie um gráfico mostrando a distribuição de tipos de quartos em diferentes bairros.

````python
# Sua resposta aqui
import matplotlib.pyplot as plt
import seaborn as sns

df.groupby(['vizinhanca', 'tipo_quarto']).size().unstack().plot(kind='bar', stacked=True, figsize=(10,7))
plt.title('Distribuição de Tipos de Quartos por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Número de Imóveis')
plt.show()
````

````python
# Crie um gráfico mostrando a distribuição de tipos de quartos em diferentes bairros.
# Sua resposta aqui
df.groupby(['grupo_vizinhanca', 'tipo_quarto']).size().unstack().plot(kind='bar', stacked=True, figsize=(10,7))
plt.title('Distribuição de Tipos de Quartos por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Número de Imóveis')
plt.show()
````

##### Tarefa: Crie um mapa, separando as cores por bairro

````python
sns.scatterplot(x="longitude",y="latitude",data=df,hue="grupo_vizinhanca")
````

````python
import geopandas as gpd
from geodatasets import get_path

gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df['longitude'], df['latitude'], crs="EPSG:4326"  # Define o sistema de coordenadas (WGS84)
))

# Carregar dados geográficos da área de Nova York
nybb = gpd.read_file(get_path('nybb'))

# Reprojetar o mapa de Nova York para o mesmo sistema de coordenadas dos pontos
nybb = nybb.to_crs(gdf.crs)

# Criando a figura e o eixo
fig, ax = plt.subplots(figsize=(10, 10))

# Plotar o mapa base de Nova York
nybb.plot(ax=ax, color='lightgrey')

# Usando Seaborn para plotar os pontos dos anúncios
sns.scatterplot(
    x=gdf.geometry.x,
    y=gdf.geometry.y,
    hue=gdf['grupo_vizinhanca'],
    palette='Set2',
    s=40,
    alpha=0.2,
    ax=ax
)

# Título e exibição
plt.title('Mapa de Anúncios do Airbnb em Nova York com Informações Detalhadas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
````

````python
sns.scatterplot(x="longitude", y="latitude", data=df, hue="preco", palette="RdBu_r", alpha=0.3, size=0.1)
````

### Questão 6: Análise Temporal
##### Tarefa: Como a quantidade de reviews por mês variou ao longo do tempo? Crie um gráfico de linha para mostrar essa variação.

````python
# Criando um grafico de linha para mostrar a variacao de reviews por mes ao longo do tempo
df['data'] = pd.to_datetime(df['ultima_review'])
df['mes'] = df['data'].dt.month
df['ano'] = df['data'].dt.year

df.groupby(['ano', 'mes'])['numero_reviews'].sum().plot(kind='line', figsize=(12, 6))
plt.title('Número de Reviews por Mês ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Número de Reviews')
plt.show()
````

### Questão 7: Comparação de Preços
##### Tarefa: Compare o preço médio dos imóveis entre os cinco bairros com mais listagens. Crie um gráfico de barras para ilustrar essa comparação.

````python
# Sua resposta aqui
top_neighbourhoods = df['vizinhanca'].value_counts().head(5).index
df_top_neighbourhoods = df[df['vizinhanca'].isin(top_neighbourhoods)]

df_top_neighbourhoods.groupby('vizinhanca')['preco'].mean().plot(kind='bar', figsize=(10,7))
plt.title('Preço Médio por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Preço Médio')
plt.show()
````
