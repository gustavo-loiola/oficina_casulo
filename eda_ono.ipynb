# Outlier

```python
df["price"].hist()
```

```python
iq1=df["price"].quantile(0.25)
iq3=df["price"].quantile(0.75)
iqr=iq3-iq1
outlier_preco=iq3+1.5*iqr
```

```python
preco_caro=df[df["price"]>outlier_preco]
```

```python
len(preco_caro)
```

```python
df=df[df["price"]<=outlier_preco]
df["price"].hist()
```

# AnÃ¡lise do preÃ§o pelo room/type

```python
for tipo in df["room_type"].unique():
    media=df.loc[df["room_type"]==tipo,"price"].mean()
    df.loc[df["room_type"]==tipo,"media_preco/room_type"]=media

df

    
```

```python
sns.barplot(x="room_type",y="media_preco/room_type",data=df)
plt.xlabel("Tipo de acomodaÃ§Ã£o")
plt.ylabel("PreÃ§o mÃ©dio")
plt.title("MÃ©dia de preÃ§o por tipo de acomodaÃ§Ã£o")
plt.show()
plt.show()
```
